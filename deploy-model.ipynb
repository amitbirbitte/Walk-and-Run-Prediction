## Load Model

```python
import pickle

model_name = "walkrun_model.pkl"
```

```python
#load the model
load_model = pickle.load(open(model_name, 'rb'))
print("Model Loaded Successfully")
```

```python
#predicting
import pandas as pd

cols = [ 'wrist', 'acceleration_x',
       'acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y', 'gyro_z']

values = [[0, 0.3031, -0.8125, 0.0888, 0.1199, -0.4099, -2.9336]]

predict = pd.DataFrame(values, columns=cols)

load_model.predict(predict)
```
